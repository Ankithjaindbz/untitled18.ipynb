```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR

# Load the dataset
data = pd.read_csv('/content/supply_chain_data (1).csv')  # Update with the correct path

# Define features and target
X = data[['Product type', 'Customer demographics', 'Revenue generated']]
y = data['Number of products sold']

# Preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Revenue generated']),
        ('cat', OneHotEncoder(), ['Product type', 'Customer demographics'])
    ])

# KMeans clustering pipeline
kmeans_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('clusterer', KMeans(n_clusters=3, random_state=42))
])

# Perform clustering
data['Cluster'] = kmeans_pipeline.fit_predict(X)

# Calculate BIC scores for each model in each cluster
models = {
    'Random Forest': RandomForestRegressor(),
    'Linear Regression': LinearRegression(),
    'SVM': SVR(kernel='linear')
}

bic_scores = {}
for cluster in data['Cluster'].unique():
    cluster_data = data[data['Cluster'] == cluster]
    X_cluster = cluster_data[['Revenue generated']]
    y_cluster = cluster_data['Number of products sold']
    bic_scores[cluster] = {}
    for model_name, model in models.items():
        model.fit(X_cluster, y_cluster)
        log_likelihood = model.score(X_cluster, y_cluster)
        if model_name == 'Linear Regression':
            n_params = len(model.coef_) + 1  # number of parameters including intercept
        elif model_name == 'Random Forest':
            n_params = X_cluster.shape[1] + 1  # number of features + 1 for intercept
        else:
            n_params = len(model.support_vectors_) + len(model.dual_coef_) + 1  # number of support vectors + number of coefficients + 1 for intercept
        n_samples = len(cluster_data)
        bic = -2 * log_likelihood + n_params * np.log(n_samples)
        bic_scores[cluster][model_name] = bic

# Print BIC scores for each model in each cluster
for cluster, models in bic_scores.items():
    print(f"Cluster {cluster}:")
    for model, bic in models.items():
        print(f"{model}: BIC = {bic:.2f}")
    print()

```

```python
import numpy as np

# BIC scores for each model across all clusters
bic_scores = {
    'Random Forest': [274.99, 276.90, 274.861],
    'Linear Regression': [634.95, 485.94, 356.76],
    'SVM': [638.03, 488.15, 359.70]
}

# Calculate posterior probabilities for each model
posterior_probabilities = {}
for model, bic_values in bic_scores.items():
    exp_bic = np.exp(-0.5 * np.array(bic_values))
    posterior_prob = exp_bic / np.sum(exp_bic)
    posterior_probabilities[model] = posterior_prob

# Print posterior probabilities for each model
for model, prob in posterior_probabilities.items():
    print(f"Model: {model}")
    for cluster, posterior in zip(['C0', 'C1', 'C2'], prob):
        print(f"Cluster {cluster}: Posterior Probability = {posterior:.4f}")
    print()
```

```python
# Perform BMA for each cluster
bma_predictions = {}
for cluster in data['Cluster'].unique():
    cluster_data = data[data['Cluster'] == cluster]
    X_cluster = cluster_data[['Revenue generated']]
    y_cluster = cluster_data['Number of products sold']
    bma_predictions[cluster] = {}
    for model_name, model in models.items():
        model.fit(X_cluster, y_cluster)
        prediction = model.predict(X_cluster)
        bma_predictions[cluster][model_name] = prediction

# Weight the predictions by the posterior probabilities
for cluster, predictions in bma_predictions.items():
    if cluster in posterior_probabilities:  # Check if posterior probabilities exist for the cluster
        posterior_probs = posterior_probabilities[cluster]
        bma_prediction = np.zeros_like(predictions[list(predictions.keys())[0]])  # Initialize BMA prediction array
        for model_name, prediction in predictions.items():
            bma_prediction += prediction * posterior_probs[model_name]  # Weighted sum of predictions
        bma_predictions[cluster]['BMA'] = bma_prediction  # Add BMA prediction to the dictionary

# Print BMA predictions for each cluster
for cluster, predictions in bma_predictions.items():
    print(f"Cluster {cluster}:")
    for model_name, prediction in predictions.items():
        print(f"{model_name}: {prediction}")
    print()
```
